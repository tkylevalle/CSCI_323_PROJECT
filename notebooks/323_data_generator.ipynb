```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)
num_records = 1500

# Top 20 UAE Universities (Inconsistent for cleaning)
uni_list = ["UOWD", "University of Wollongong Dubai", "uowd", "AUS", "Am. Uni Sharjah",
            "Khalifa Uni", "KU", "Heriot-Watt", "Middlesex Dubai", "AUD", "CUD", "RIT Dubai",
            "Birmingham Uni", "Zayed Uni", "Manipal", "Amity", "Murdoch", "ADU", "Uni of Dubai", "BUiD"]

data = []
for i in range(num_records):
    # Introduce Messy Data (Nulls & Noise)
    math = np.random.choice([np.nan, np.random.normal(70, 15)], p=[0.05, 0.95])
    physics = np.random.normal(65, 20)
    english = np.random.normal(75, 10)
    ict = np.random.normal(78, 12)
    emsat_eng = np.random.normal(1400, 250)

    # Logic for synthetic "Ground Truth" (AI will learn this hidden pattern)
    avg = (np.nan_to_num(math) + physics + english + ict) / 4
    if avg > 80 and emsat_eng > 1550:
        actual_path = "Year 13"
    elif avg < 60 or emsat_eng < 1100:
        actual_path = "Foundation"
    else:
        actual_path = random.choice(["Year 13", "Foundation"]) # Noise

    data.append([
        f"ID_{random.randint(1000, 9999)}",
        random.choice(["Male", "female", "M", "F", "FEMALE"]), # Messy Gender
        round(math, 1) if not pd.isna(math) else np.nan,
        round(physics, 1), round(english, 1), round(ict, 1), round(emsat_eng, 0),
        random.choice(uni_list),
        actual_path
    ])

df = pd.DataFrame(data, columns=["Student_Ref", "Gender", "Math", "Phys", "Eng", "ICT", "EmSAT_Eng", "Target_Uni", "Pathway"])
df = pd.concat([df, df.sample(n=100)], ignore_index=True) # Duplicates
df.to_csv("RAW_EduPath_Data.csv", index=False)
print("RAW_EduPath_Data.csv generated with duplicates and nulls.")
```
